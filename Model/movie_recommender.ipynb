```python
import pandas as pd
import numpy as np
```

```python
movies= pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
movies=movies.merge(credits,on="title")
```

```python
movies=movies[["title","genres","movie_id","keywords","overview","tagline","cast","crew"]]
```

```python
movies["genres"]
```

```python
movies.dropna()
```

```python
import ast
ast.literal_eval
```

```python
def convert(x):
    L=[]
    for i in ast.literal_eval(x):
        L.append(i["name"])
    return L
```

```python
movies["genres"]=movies["genres"].apply(convert)
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
def convertx(x):
    L=[]
    c=0
    for i in ast.literal_eval(x):
        if c!=3:
            L.append(i["name"])
        else:
            break
        c+=1
    return L
```

```python
movies["cast"]=movies["cast"].apply(convertx)
```

```python
movies
```

```python
def convertd(x):
    L=[]
    for i in ast.literal_eval(x):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"]=movies["crew"].apply(convertd)
```

```python
movies["tagline"]=movies["tagline"].fillna("unknown")

```

```python
movies["overview"] = movies["overview"].apply(lambda x: x.split() if isinstance(x, str) else [])
```

```python
movies["tagline"]= movies["tagline"].apply(lambda x: x.split() if isinstance(x,str) else [])
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"]=movies["genres"]+movies["keywords"]+movies["overview"]+movies["tagline"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
df=movies[["title","movie_id","tags"]]
```

```python
df["tags"]=df["tags"].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df["tags"]=df["tags"].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(x):
    l=[]
    for i in x.split():
        l.append(ps.stem(i))
    return " ".join(l)
        
```

```python
df.head()
```

```python
df["tags"]=df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(df["tags"]).toarray()
```

```python
len(list(cv.get_feature_names_out()))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommand(movie):
    movie_index = df[df["title"]==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].title)
    
```

```python
recommand("Batman Begins")
```
